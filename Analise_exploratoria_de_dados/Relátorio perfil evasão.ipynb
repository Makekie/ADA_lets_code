# Introdução

    O banco *XXXX* solicitou uma análise de um de seus bancos de dados para tentar caracterizar os clientes que estão deixando a empresa. A base de dados, contendo 10000 linhas, tem como principais informações de cada cliente: a pontuação de crédito, país de moradia, gênero, idade, confiança, saldo, número de produtos, posse de cartão de crédito, é ativo, deposito estimado e saida do banco.

# Levantamento de hipóteses e estudo dos dados

    Antes de começar uma Análise Exploratória da base de Dados (AED) foi necessário conferir os dados e selecionar os mais importantes. Com isso, não foi observado inconsistência nos dados sendo feito apenas a exclusão das colunas relacionadas ao sobrenome e identificação do consumidor. 

```python
import pandas as pd
import seaborn as sns
ds_churn = pd.read_csv("churn_relatorio.csv")
```

1. O primeiro questionamento surge ao pensar nas idades dos clientes, pelo banco ter clientes desde os 18 até os 84 anos, cada idade deve ter um comportamento diferente?

    O histograma da idade mostra uma grande concentração de clientes entre 30 e 50 anos. No entanto, é possível tirar pouca informação dele (como por exemplo ele nos da a entender que, proporcionalmente, entre 47 e 60 anos é a idade em que mais tem saída que permanência). Já observando para a impressão de um gráfico boxplot, fica mais claro a ideia que as pessoas mais velhas tendem a sair enquanto as mais novas não.

```python
sns.displot(ds_churn, x = 'Age', kde = True, hue = 'Exited')
```

```python
sns.catplot(data = ds_churn, y = "Age", x = "Exited", kind = 'box')
```

2. E se ao invés de observar as idades separadamente fosse feito um agrupamento delas? Então surge a ideia de utilizar o conceito de gerações. Como as diferentes gerações se comportam em relação a saída do banco? Para poder análisar tal conduta foi criada uma nova coluna para apontar a geração que cada pessoa pertence e para definir quando cada geração começa e termina utilizou-se da seguinte fonte: [Definição das gerações](https://beieducacao.com.br/geracoes-x-y-z-e-alfa-como-cada-uma-se-comporta-e-aprende/).

```python
ds_churn[["Age", "Generations"]].head()
```

    Ao estudar esse agrupamento percebe-se que, tanto proporcionalmente quanto absolutamente, o número de clientes que deixam o banco é maior para as pessoas pertencentes a geração X (nascidos entre 1965-1980). 

```python
sns.displot(ds_churn, x = 'Generations', hue = 'Exited')
```

3. Outro dado que ajudou a caracterizar o perfil daqueles que deixam o banco foi sobre o gênero de cada cliente. Será que existe uma movimento de saída maior entre homens e mulheres? A comparação no número de mulheres que saem em relação a homens, mostra uma diferença, desfavorável as mulheres, de 241 pessoas. Esse comportamento já demonstra uma maior probabilidade do público feminino deixar o banco, pois é importante salientar que o número de homens, no total, é maior que o das mulheres.

```python
sns.displot(ds_churn, x = "Gender", hue = "Exited")
```

    Agora estudando o comportamento da idade juntamente com a diferença entre gêneros começa a ficar mais nítido o perfil do cliente que deixa do banco e demonstra uma relação entre os dois tópicos. Até então o perfil é composto pelas mulheres da geração X.

```python
sns.displot(ds_churn, x = 'Gender', col = "Generations", hue = 'Exited')
```

4. A posição geografica do cliente deve ser um fator importante na caracterização do perfil? Com esse quarto questionamento é possível observar que os alemães, proporcionalmente, são os que mais saem do banco. Adotando a mesma linha de raciocíneo usada acima e relacionando o lugar com a idade e o gênero, o perfil fica mais detalhado. As mulheres da geração X da alemanha são as que mais evadem.

```python
sns.displot(ds_churn, x = "Geography", hue = "Exited")
```

```python
sns.displot(ds_churn, x = 'Generations', col = 'Geography', hue = 'Exited')
```

```python
sns.displot(ds_churn, x = 'Gender', col = 'Geography', hue = 'Exited')
```

5. Por fim, estudando as possíveis correlações entre as colunas da base de dados, para procurar alguma relação não abordada que seja influente, tem-se uma correlação positiva de 0.12 entre o Saldo e a Saída de clientes.

```python
sns.heatmap(ds_churn.corr(), annot = True, fmt = '.2f', linewidth = 1, cmap = "Spectral")
```

    Apesar de não ser tão distinta, o saldo em conta demonstra que aqueles com maior valor depositado tendem a sair. No entanto não consegui segregar os dados de forma significativa (como fazer uma divisão entre classes A, B, C, D e E). 

```python
sns.catplot(data = ds_churn, x = 'Exited', y = "Balance", kind = 'box')
```
