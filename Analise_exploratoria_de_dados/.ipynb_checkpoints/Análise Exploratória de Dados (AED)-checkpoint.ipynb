Hipóteses levantadas:
Quanto a idade influênciou na saída

```python
#Olhando a idade das pessoas que saíram
ds_churn.query("Exited == 1")["Age"].value_counts()
```

```python
geracoes = [0, 11, 25, 41, 76, 100]
#Alfa, Z, Y, X, Baby Boomers, segundo link publicado em 2021: https://beieducacao.com.br/geracoes-x-y-z-e-alfa-como-cada-uma-se-comporta-e-aprende/
#Os valores limites de idade foram adaptados para esse ano (2022)
```

```python
#Olhando a idade das pessoas que sairam, só que separando entre gerações
ds_churn.query("Exited == 1")["Age"].value_counts(bins = geracoes)
```

```python
ds_churn['Generations'] = pd.cut(ds_churn.Age,
                                bins = geracoes,
                                labels = ["Alfa", "Z", "Y", "X", "Boomer"],
                                right = True,
                                include_lowest = True)
ds_churn
```

```python
sns.displot(ds_churn, x = 'Age', kde = True)
```

```python
sns.displot(ds_churn, x = 'Age', col = 'Gender', kde = True)
```

```python
sns.displot(ds_churn, x = 'Age', col = 'Gender', kde = True, hue = 'Exited')
```

```python
sns.displot(ds_churn, x = 'Age', col = 'Geography', kde = True)
```

```python
sns.displot(ds_churn, x = 'Age', col = 'Geography', kde = True, hue = 'Exited')
```

```python
sns.displot(ds_churn, x = 'Age', col = 'Exited', kind = 'kde', hue = 'Tenure')
```

```python
sns.catplot(data = ds_churn, y = "Age", x = "Gender", kind = 'box', hue = 'Exited')
```

```python
sns.displot(ds_churn, x = 'Generations')
```

```python
sns.displot(ds_churn, x = 'Generations', hue = 'Exited')
```
